```python
y_train = data_train['is_fraud']
y_test = data_test['is_fraud']

data_train.drop(columns=['dob','is_fraud'],inplace=True)
data_test.drop(columns = ['dob','is_fraud'],inplace=True)
```

```python
data_train['trans_year'] = data_train['trans_date_trans_time'].dt.year
data_train['trans_month'] = data_train['trans_date_trans_time'].dt.month
data_train['trans_day'] = data_train['trans_date_trans_time'].dt.day
data_train['trans_hour'] = data_train['trans_date_trans_time'].dt.hour
data_train['trans_minute'] = data_train['trans_date_trans_time'].dt.minute
data_train.drop(columns=['trans_date_trans_time'], inplace=True)
```

```python
data_test['trans_year'] = data_test['trans_date_trans_time'].dt.year
data_test['trans_month'] = data_test['trans_date_trans_time'].dt.month
data_test['trans_day'] = data_test['trans_date_trans_time'].dt.day
data_test['trans_hour'] = data_test['trans_date_trans_time'].dt.hour
data_test['trans_minute'] = data_test['trans_date_trans_time'].dt.minute
data_test.drop(columns=['trans_date_trans_time'], inplace=True)
```

```python
data_train_gender = pd.get_dummies(data_train['gender'])
data_train_gender
```

```python
data_test_gender = pd.get_dummies(data_test['gender'])
data_test_gender
```

```python
data_train = pd.concat([data_train, data_train_gender], axis=1)

data_train.drop(columns=['gender'], inplace=True)
```

```python
data_test = pd.concat([data_test, data_test_gender], axis=1)

data_test.drop(columns=['gender'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

categorical_cols = ['merchant', 'category', 'first', 'last', 'street', 'city', 'state', 'zip', 'F', 'M','job']

label_encoder = LabelEncoder()

for col in categorical_cols:
    data_train[col] = label_encoder.fit_transform(data_train[col])

scaler = StandardScaler()
numerical_cols = ['cc_num', 'amt', 'lat', 'long', 'city_pop', 'merch_lat', 'merch_long', 'age', 'trans_year', 'trans_month', 'trans_day', 'trans_hour', 'trans_minute']
data_train[numerical_cols] = scaler.fit_transform(data_train[numerical_cols])
```

```python

categorical_cols = ['merchant', 'category', 'first', 'last', 'street', 'city', 'state', 'zip', 'F', 'M','job']

label_encoder1 = LabelEncoder()

for col in categorical_cols:
    data_test[col] = label_encoder1.fit_transform(data_test[col])

scaler = StandardScaler()
numerical_cols = ['cc_num', 'amt', 'lat', 'long', 'city_pop', 'merch_lat', 'merch_long', 'age', 'trans_year', 'trans_month', 'trans_day', 'trans_hour', 'trans_minute']
data_test[numerical_cols] = scaler.fit_transform(data_test[numerical_cols])
```

# Xgboost 

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(n_estimators=200, learning_rate=0.05, max_depth=7, random_state=2, n_jobs=-1)
xgb_model.fit(data_train, y_train)

y_pred_xgb = xgb_model.predict(data_test)
print("XGBoost with scale_pos_weight:")
print(classification_report(y_test, y_pred_xgb))
```

```python
from sklearn.metrics import log_loss
loss_score = log_loss(y_test, y_pred_xgb)
print("Log Loss:", loss_score)
```

```python
accuracy_score(y_pred_xgb,y_test)
```

# Easy Ensemble Classifier

```python
from imblearn.ensemble import EasyEnsembleClassifier

eec = EasyEnsembleClassifier(random_state=42)

eec.fit(data_train, y_train)

y_pred_eec = eec.predict(data_test)

print("EasyEnsemble:")
print(classification_report(y_test, y_pred_eec))

loss_score = log_loss(y_test, y_pred_eec)
print("Log Loss:", loss_score)


print("Accuracy socre : ",accuracy_score(y_pred_eec,y_test))
```

# Balanced Bagging Classifier

```python
from imblearn.ensemble import BalancedBaggingClassifier

bbc = BalancedBaggingClassifier(n_estimators=100, random_state=2, n_jobs=-1)
bbc.fit(data_train, y_train)

y_pred_bbc = bbc.predict(data_test)

print("Classification Report :",classification_report(y_test, y_pred_bbc))


loss_score = log_loss(y_test, y_pred_bbc)
print("Log Loss:", loss_score)


print("Accuracy socre : ",accuracy_score(y_pred_bbc,y_test))
```

# Balanced Random Forest Classifier

```python
from imblearn.ensemble import BalancedRandomForestClassifier

brf = BalancedRandomForestClassifier(random_state=42)

brf.fit(data_train, y_train)


```

```python
y_pred_brf = brf.predict(data_test)
print("Classification Report :",classification_report(y_test, y_pred_brf))

loss_score = log_loss(y_test, y_pred_brf)
print("Log Loss:", loss_score)

print("Accuracy socre : ",accuracy_score(y_pred_brf,y_test))
```
